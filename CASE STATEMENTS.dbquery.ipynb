```sql
CREATE OR REPLACE TABLE employees AS
SELECT * FROM VALUES
(1, 'John', 'IT', 85000),
(2, 'Sara', 'HR', 60000),
(3, 'Mark', 'IT', 75000),
(4,'Lucy', 'Finance', 95000),
(5, 'Tom', 'HR', 55000)
AS t(emp_id, emp_name, department, salary);

SELECT *
FROM employees;

SELECT emp_name,
       department,
       salary,
       CASE 
       WHEN department = 'IT' AND salary > 80000 THEN 'Senior IT'
       WHEN department = 'HR' AND salary > 55000 THEN 'Experienced'
       ELSE 'Staff'
       END AS postion_level
       FROM employees;



       CREATE OR REPLACE TABLE students AS
SELECT * FROM VALUES
(1, 'Ana', 92),
(2, 'Ben', 76),
(3, 'Cara', 59),
(4,'David', 83),
(5, 'Ella', 68)
AS students(students_id, students_name, score);

SELECT *
FROM students;

SELECT students_name,
       score,
       CASE
          WHEN score >= 90 THEN 'A'
          WHEN score BETWEEN 80 AND 89 THEN 'B'
          WHEN score BETWEEN 70 AND 79 THEN 'C'
          WHEN score BETWEEN 60 AND 69 THEN 'D'
          WHEN score < 60 THEN 'Fail'
          END AS Grade
          FROM students;


                 CREATE OR REPLACE TABLE deliveries AS
SELECT * FROM VALUES
(1, 45),
(2, 80),
(3, 30),
(4,65) ,
(5, 100)
AS t(deliveries_id, delivery_time_minutes);

SELECT deliveries_ID,
        delivery_time_minutes,
        CASE
          WHEN delivery_time_minutes <= 30 THEN 'Fast'
          WHEN delivery_time_minutes BETWEEN 30 AND 60 THEN 'On Time'
          WHEN delivery_time_minutes > 60 THEN 'Late'
          END AS Perfomance
          FROM deliveries;

  
                 CREATE OR REPLACE TABLE Customers AS
SELECT * FROM VALUES
(101, 'Alice', 450.00, 12),
(102,'Bob', 1200.00, 8),
(103, 'Charlie', 750.50, 25),
(104, 'Diana', 2300.00, 5) ,
(105, 'Ethan', 180.75, 3)
AS t(customer_id, customer_name,  total_spend, purchase_count);     

SELECT customer_name,
       total_spend,
       purchase_count,
       CASE
          WHEN total_spend >= 2000 OR purchase_count >= 20 THEN 'Platinum'
          WHEN total_spend >= 1000 OR purchase_count >= 10 THEN 'Gold'
          ELSE 'Silver' 
          END AS Label
FROM Customers;


CREATE OR REPLACE TABLE Products AS
SELECT * FROM VALUES
('P001','Wireless Earbuds', 1250 , 18750.00, 0.08),
('P002','Smart Watch', 890 , 44500.00, 0.12),
('P003', 'Laptop Sleeve', 3200, 12800.00, 0.03),
('P004', 'Bluetooth Speaker', 650, 9750.00, 0.15) ,
('P005', 'USB-C Cable', 4500, 6750.00, 0.05)
AS t(product_id, product_name, units_sold, revenue, return_rate);  

SELECT product_name, 
       units_sold,
       revenue, 
       return_rate,
       CASE
          WHEN revenue >= 20000  THEN 'Top Perfomer'
          WHEN units_sold >= 1000 AND return_rate <0.10 THEN 'Strong Seller'
          ELSE 'Needs Attention' 
          END AS Label
FROM products;



CREATE OR REPLACE TABLE Subscribers AS
SELECT * FROM VALUES
('S001','user1@example.com', 0.65  , 0.12, 3),
('S002','user2@example.com', 0.42, 0.08, 45),
('S003', 'user3@example.com', 0.81, 0.25, 1),
('S004', 'user4@example.com',0.55, 0.55, 12) ,
('S005', 'user5@example.com', 0.33, 0.03, 90)
AS t(subscriber_id, email, open_rate, click_rate, last_open_days_ago);

select email, open_rate, click_rate, last_open_days_ago
from subscribers;

SELECT email, open_rate, click_rate, last_open_days_ago,
    CASE
        WHEN open_rate >= 0.70 AND click_rate >= 0.15 THEN 'Highly Engaged'
        WHEN open_rate >= 0.50 OR last_open_days_ago <= 14 THEN 'Moderately Engaged'
        ELSE 'Low Engagement'
    END AS Label
FROM subscribers;

```
